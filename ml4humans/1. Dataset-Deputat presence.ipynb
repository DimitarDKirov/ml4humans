# Depustats Dataset

The dataset used in the following notebooks will revolve around the following problem: Predict whether a given member of parliament(a deputat) will come to the parliament, given his past behaviour.   
The data was gathered from https://www.parliament.bg/bg/plenaryst/ns/52/ID/10290 in the form of excel files, cleaned and parsed.    
As a result, for each deputat we have his daily work presence for the period 2019-01-11-2020-04-03. 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set('notebook')
sns.set_style('white')

data = pd.read_csv('../data/deputat_daily_features.csv')
data.head()
```

```python
fig, ax = plt.subplots(figsize=(16,2))
sns.despine(ax=ax)

deputat = np.random.choice(data.deputat.unique())
ax.set_title(f"Присъствие на {deputat}",loc='left', size=16)
data.loc[data.deputat==deputat].set_index('date')['present'].astype(int).plot(style='bo', ax=ax);
```

Additionally, I have added the following features: was the deputat present in the last assembly? What proportion of the last 5 assemblies was he present in?
