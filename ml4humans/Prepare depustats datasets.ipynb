```python
import pandas as pd
from pathlib import Path
import re
from dateutil import parser
```

```python
def extract_date_from_fn(fn="/home/martin/data/parliament2019-2020/20191217103811iv061219.xls"):
    return parser.parse(re.findall(".*iv(\d+).xls", fn)[0], dayfirst=True)
extract_date_from_fn()
```

```python
path = Path("/home/martin/data/parliament2019-2020")

date_dfs = []
for fn in path.glob("*iv*.xls"):
    date = extract_date_from_fn(str(fn))
    contents = pd.read_excel(str(fn)).iloc[1:]
    contents = contents[contents.columns[:5]]
    contents = contents.rename(columns={contents.columns[0]: "deputat",  'Unnamed: 3': 'party', 'Unnamed: 4': 'present' })
    contents = contents[['deputat', 'party', 'present']]
    contents['present'] = contents['present'] == 'П'
    if contents['present'].sum() < 120:
        display(fn, contents['present'].sum())
        display(pd.read_excel(str(fn)).iloc[1:])
    contents['date'] = date
    date_dfs.append(contents)
    if len(contents) < 239:
        display(date)
        display(contents)
data = pd.concat(date_dfs)
```

```python
data.to_csv(str(path / 'cleaned.csv'))
```
